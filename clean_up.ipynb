```python
#remove prefixes
name_list=df['student_name'].tolist()
for name in name_list:
    print(name.split(), len(name.split()))
```

```python
#separate names with 3 
fixer_uppers=[]
for name in name_list:
    if len(name.split()) >= 3:
        fixer_uppers.append(name)
len(fixer_uppers)
```

```python
#longest prefix is "Miss"
#print(fixer_uppers)
```

```python
#isolate prefixes "Miss" is longest at 4 characters. 
prefix_list=[]
for name in fixer_uppers:
    if len(name.split()[0]) <= 4:
        prefix_list.append(name.split()[0])
print(prefix_list)
```

```python
#do the same for suffixes "[-1]""
suffix_list=[]
for name in fixer_uppers:
    if len(name.split()[-1]) <= 3:
        suffix_list.append(name.split()[-1])
print(suffix_list)
```

```python

#isolate prefixes "Miss" is longest at 4 characters. 
prefix_list=[]
for name in fixer_uppers:
    if len(name.split()[0]) <= 4:
        prefix_list.append(name.split()[0])
print(prefix_list)
```

```python
#create a set of unique items to isolate what we need to delete

set(suffix_list)
```

```python
set(prefix_list)
```

```python
#get rid of: 'Dr.', 'Mr.', 'Ms.', 'Miss', 'MD', 'DDS', 'DVM', 'Ph.D.' from original list of student names along with the white space ""
```

```python
remove=['Dr. ', 'Mr. ', 'Ms. ', 'Miss ', 'MD', 'DDS', 'DVM', 'Ph.D.']
for x in remove:
    df['student_name']=df['student_name'].str.replace(x, "")
#validate
# print(df['student_name'].tolist())
fixed_names=df['student_name'].tolist()
print(fixed_names)
```

```python
#create new column with no prefix/suffix in names
clean_names = []

for name in fixed_names:
    if len(name.split()) >= 3:
        clean_names.append(name)
len(clean_names)
```
